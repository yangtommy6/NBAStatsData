```python
import pandas as pd
import matplotlib.pyplot as plt

# Read data
data = pd.read_csv("NBA2013_14Stats.csv")

# Calculate TS%
data["TS%"] = data["PTS"] / (2 * (data["FGA"] + 0.44 * data["FTA"]))

# Filter top 15 scoring leaders
top_15 = data.nlargest(15, "PTS")
top_15_filtered = top_15[["PLAYER", "PTS", "TS%"]]

# Create the line chart
fig, ax = plt.subplots(figsize=(10, 6)) # increase figure size
top_15_filtered.plot(x="PLAYER", y="PTS", ax=ax, marker="o", label="Points per Game")
top_15_filtered.plot(x="PLAYER", y="TS%", ax=ax, marker="s", secondary_y=True, label="True Shooting Percentage")

# Set plot title and axis labels
ax.set_title("Top 15 Players by Points and True Shooting Percentage")
ax.set_xlabel("Players")
ax.set_ylabel("Points per Game")
ax.right_ax.set_ylabel("True Shooting Percentage")

# Customize plot appearance
ax.legend(loc="upper left")
ax.right_ax.legend(loc="upper right")
plt.xticks(rotation=30, ha="right") # rotate x-tick labels
plt.subplots_adjust(bottom=0.15) # adjust bottom margin to fit x-tick labels
plt.show()
```
